```python
# To use pandas we need to import first.
# Pandas will represent data in 2 different structures:
# Series - 1D labeled array
# DataFrame - 2D labeled data structure with columns of potentially different types

import pandas as pd

df = pd.DataFrame({
    'emp_id' : [1, 2, 3, 4],
    'emp_name' : ['Alice', 'Bob', 'Charlie', 'David'],
    'emp_salary' : [70000, 80000, 90000, 60000],
    'department' : ['HR', 'Finance', 'IT', 'Marketing']
})

# To pring the DataFrame we just created, we can simply type its name
df

#print(df)
```

```python
# In a DF each column is a Series.
# From abobe DataFrame we can extract a column as Series like below:

emp_names = df['emp_name']

print(emp_names)

```

```python
# We can create a new series like below in an existing DataFrame.

gender = pd.Series(['M', 'F', 'M', 'F'], name='gender')

print(gender)

df['gender'] = gender

df

# Another way to create a Series and add to existing DataFrame
df['experience'] = pd.Series([5, 7, 6, 4])

df

# To add a series to specific index location
df.insert(2, 'status', pd.Series(['Active', 'Active', 'Inactive']))

df
```

```python
# Data Frame describe() will give statistical summary of numerical columns.

df.describe()
```

```python
# To get the max value of a Numerical column
print(df)

max_salary = df['emp_salary'].max()
print("Max Salary:", max_salary)

# To get the min value of a Numerical column

min_salary = df['emp_salary'].min()
print("Min Salary:", min_salary)
```
